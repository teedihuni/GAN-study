### numpy with python

```python
import numpy as np
import torch
```

```python
size = 50

a = np.random.rand(size, size)
b = np.random.rand(size, size)
```

```python
%%timeit

x = np.dot(a,b)
```

numpy 쓰지 않고 행렬 계산

```python
%%timeit

c = np.zeros((size,size))

for i in range(size):
  for j in range(size):
    for k in range(size):
      c[i,j] += a[i,k] * b[k,j]
    pass
  pass
```

### Tensors on GPU

```python
x = torch.cuda.FloatTensor([3.5])
x.type()
```

```python
x.device
```

device에 올린 값은 GPU로 계산하게 됨

```python
y= x*x
print(y)
```

처음에 했던 50차원의 행렬 곱을 GPU로 계산

```python
aa = torch.cuda.FloatTensor(a)
bb = torch.cuda.FloatTensor(b)
```

```python
%%timeit

cc = torch.matmul(aa, bb)
```
